```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import *
```

```python
data = data = pd.read_csv('/Users/panbingqing/Downloads/Python 系列用到的数据/data for matplotlib/S&P500.csv',
                  index_col=0,
                  parse_dates=True,
                  dayfirst=True)
data.head()
```

- index_col = 0 是说把第一列 Date 当成行标签 (index)
- parse_dates = True 是说把行标签转成 date 对象
- dayFirst = True 是说日期是 DD/MM/YYYY 这样的格式

```python
spx = data['Adj Close'].loc['2007-01-01':'2010-01-01']
spx.sample(5)
```

```python
plt.plot(spx.values)
# 在 plot() 函数里面只有变量 y 时 (y = spx.values)，那么自变量就是默认赋值为 range(len(y))。
# 此外我们没有设置图的尺寸，像素、线的颜色宽度、坐标轴的刻度和标签、图例、标题等等，
# 所有设置都用的是 matplotlib 的默认设置。
```

```python
plt.rcParams #查看默认属性
```

```python
#Echarts 中的颜色
r_hex = '#dc2624'     # red,       RGB = 220,38,36
dt_hex = '#2b4750'    # dark teal, RGB = 43,71,80
tl_hex = '#45a0a2'    # teal,      RGB = 69,160,162
r1_hex = '#e87a59'    # red,       RGB = 232,122,89
tl1_hex = '#7dcaa9'   # teal,      RGB = 125,202,169
g_hex = '#649E7D'     # green,     RGB = 100,158,125
o_hex = '#dc8018'     # orange,    RGB = 220,128,24
tn_hex = '#C89F91'    # tan,       RGB = 200,159,145
g50_hex = '#6c6d6c'   # grey-50,   RGB = 108,109,108
bg_hex = '#4f6268'    # blue grey, RGB = 79,98,104
g25_hex = '#c7cccf'   # grey-25,   RGB = 199,204,207
```

```python
plt.figure(figsize=(16,6),dpi=100)
plt.plot( spx.values,color=dt_hex,
          linewidth=2, linestyle='-' )
```

```python
fig = plt.figure(figsize=(16,6),dpi=100)
ax = fig.add_subplot(1,1,1)
ax.plot(spx.index, spx.values, color=dt_hex, linewidth=2, linestyle='-')
y = spx.values
ax.set_xlim(['1/1/2007', '1/1/2010'])
ax.set_ylim( y.min()*0.8,  y.max()*1.2)
```

- ax.plot()画折线
- ax.set_xlim(),ax.set_ylim()设置横轴和纵轴的边界

```python
fig = plt.figure(figsize=(16,6),dpi=100)
ax = fig.add_subplot(1,1,1)
x = spx.index
y = spx.values
ax.plot(y, color=dt_hex, linewidth=2, linestyle='-')

ax.set_xlim(-1,len(x)+1)
ax.set_ylim( y.min()*0.8,  y.max()*1.2)

ax.set_xticks(range(0,len(x),40))
ax.set_xticklabels([x[i].strftime('%Y-%m-%d') for i in ax.get_xticks()],rotation=90)
```

- 先用 ax.set_ticks() 设置出数值刻度
- 再用 ax.set_xticklabels() 在对应的数值刻度上写标签

```python
#添加图例
#添加图例 (legend) 非常简单，只需要在 ax.plot() 里多设定一个参数 label，然后用ax.legend()
#frameon = True 给图例加了外框。
fig = plt.figure(figsize=(16,6),dpi=100)
ax = fig.add_subplot(1,1,1)
x = spx.index
y = spx.values
ax.plot(y, color=dt_hex, linewidth=2, linestyle='-',label='S&P500')
ax.legend(loc=0,frameon=True)

ax.set_xlim(-1,len(x)+1)
ax.set_ylim( y.min()*0.8,  y.max()*1.2)

ax.set_xticks(range(0,len(x),40))
ax.set_xticklabels([x[i].strftime('%Y-%m-%d') for i in ax.get_xticks()],rotation=90)
```

上图问题：两条线量纲不一样，导致看起来不顺眼
- 用两个坐标系
- 用两幅图

```python
#两个坐标系
fig = plt.figure(figsize=(16,6),dpi=100)
ax1 = fig.add_subplot(1,1,1)

x = spx.index
y1 = spx.values
y2 = vix.values

ax1.plot(y1,color = dt_hex,linewidth=2,linestyle='-',label='S&P500')
ax1.set_xlim(-1,len(x)+1)
ax1.set_ylim( np.vstack([y1,y2]).min()*0.8, np.vstack([y1,y2]).max()*1.2)

x_tick=range(0,len(x),40)
x_label = [x[i].strftime('%Y-%m-%d') for i in x_tick]
ax1.set_xticks(x_tick)
ax1.set_xticklabels(x_label,rotation=90)
ax1.legend(loc='upper left',frameon=True)

#增加另一个坐标
ax2 = ax1.twinx()
ax2.plot(y2,color = r_hex,linewidth=2,linestyle='-',label='S&P500')
ax1.legend(loc='upper lright',frameon=True)
```

- 2017-10-11: 牛市顶点
- 2008-03-12: 贝尔斯登倒闭
- 2008-09-15: 雷曼兄弟倒闭
- 2009-01-20: 苏格兰皇家银行股票抛售
- 2009-04-02: G20 峰会
把这几个事件点添加在图上

```python
fig = plt.figure(figsize=(16,6),dpi=100)
crisis_data = [(datetime(2007,10,11), '牛市顶点'),
              (datetime(2008,3,12),'贝尔兄弟倒闭'),
               (datetime(2008,9,15),'雷曼兄弟倒闭'),
               (datetime(2009,1,20),'苏格兰银行抛售'),
               (datetime(2009,4,2),'G20峰会')]

ax1 = fig.add_subplot(1,1,1)

x = spx.index
y1 = spx.values
y2 = vix.values

ax1.plot(y1,color = dt_hex,linewidth=2,linestyle='-',label='S&P500')
ax1.set_xlim(-1,len(x)+1)
ax1.set_ylim( np.vstack([y1,y2]).min()*0.8, np.vstack([y1,y2]).max()*1.2)

x_tick=range(0,len(x),40)
x_label = [x[i].strftime('%Y-%m-%d') for i in x_tick]
ax1.set_xticks(x_tick)
ax1.set_xticklabels(x_label,rotation=90)
ax1.legend(loc='upper left',frameon=True)


for date,label in crisis_data:
    date = date.strftime('%Y-%m-%d')
    xi = x.get_loc(date)
    yi = spx.asof(date)
    ax1.scatter(xi,yi,80,color=r_hex)
    ax1.annotate(label,xy=(xi,yi+60),
                xytext=(xi,yi+300),
                arrowprops=dict(facecolor='black'),
                horizontalalignment='left')

#增加另一个坐标
ax2 = ax1.twinx()
ax2.plot(y2,color = r_hex,linewidth=2,linestyle='-',label='S&P500',alpha=0.3)
ax1.legend(loc='upper lright',frameon=True)
```
