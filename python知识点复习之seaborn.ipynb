```python
#鸢尾花识别
iris_data = sns.load_dataset('iris')
iris_data.sample(10)
```

```python
iris_data.describe()
```

```python
sns.pairplot(iris_data,hue='species')
#hue='species'表示根据类下不同的值赋予不同的颜色
```


- 对角线上的 4 张都是某个变量和自身的关系，用分布图 (dist plot)。

- 非对角线的 12 张就是某个变量和另一个变量的关系，用散点图 (scatter plot)。比如第一行第二列的图描述的就是萼片长度 (看纵轴第一个 sepal_length_cm 字样) 和萼片宽度 (看横轴第二个 sepal_width_cm 字样)。

```python
cond = (iris_data['species'] == 'setosa') & (iris_data['sepal_width'] < 2.5)
iris_data = iris_data[~cond] # 取出宽度大于2.5的数据
```

```python
iris_data.loc[iris_data['species'] == 'setosa','sepal_width'].hist()
```

```python
iris_data[iris_data['sepal_length'].isnull() |
iris_data['sepal_width'].isnull() |
iris_data['petal_length'].isnull() |
iris_data['petal_width'].isnull()]
```

### 线性回归
pairplot() 除了画出变量之间的关系图，通过设置里面参数 kind = 'reg'，还可在非对角图上对那些散点做线性回归。

```python
sns.pairplot( iris_data, kind='reg',hue='species' )
```
